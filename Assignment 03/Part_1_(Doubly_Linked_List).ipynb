```python
class Patient:
  def __init__(self, id, name, age, bloodgroup, next, prev):
    self.id = id
    self.name = name
    self.age = age
    self.bloodgroup = bloodgroup
    self.next = next
    self.prev = prev
```

```python
class WRM:
  def __init__(self):
    self.dh = Patient(None,None,None,None,None,None)
    self.dh.next = self.dh
    self.dh.prev = self.dh
    self.tail = self.dh

  def registerPatient(self,id, name, age, bloodgroup):
    new = Patient(id, name, age, bloodgroup, self.dh, self.tail)
    self.tail.next = new
    self.dh.prev = new
    self.tail = self.tail.next
    return "Success registering patient\n"

  def servePatient(self):
    temp = self.dh.next
    print(f"\n{temp.id} is served\n")
    temp.next.prev = self.dh
    temp.prev.next = temp.next
    self.tail = self.dh

  def showAllPatient(self):
    temp = self.dh.next
    while temp != self.dh:
      print(temp.name, end = " ")
      temp = temp.next
    print()

  def canDoctorGoHome(self):
    temp = self.dh.next
    count = 0
    while temp != self.dh:
      count += 1
      temp = temp.next
    if count > 0:
      return True
    return False

  def cancelAll(self):
    self.dh.next = self.dh
    self.dh.prev = self.dh
    self.tail = self.dh

  def ReverseTheLine(self):
    temp = self.dh.prev
    self.dh.next = temp
    while temp != self.dh:
      temp.prev, temp.next = temp.next, temp.prev
      if temp.next == self.dh:
          self.dh.prev = temp
      temp = temp.next
    return self.dh
```

```python
print("**Welcome to Waiting Room Management System**")
Main = WRM()
while True:
  print("===Choose an Options===\n1. Add Patient\n2. Serve Patient\n3. Show All Patients\n4. Can Doctor go Home?\n5. Cancel all Appointment\n6. ReverseTheLine\n7. Exit\n=======================")
  inp = int(input("Enter you choice: "))
  if inp == 1:
    print()
    id = int(input("Enter ID: "))
    name = input("Enter Name: ")
    age = int(input("Enter Age: "))
    bg = input("Enter Blood Group: ")
    add = Main.registerPatient(id, name, age, bg)
    print(add)

  elif inp == 2:
    if Main.canDoctorGoHome():
      Main.servePatient()
    else:
      print("\nNo patient to be served\n")

  elif inp == 3:
    if Main.canDoctorGoHome():
      Main.showAllPatient()
      print()
    else:
      print("\nNo patients in the WRM\n")

  elif inp == 4:
    if Main.canDoctorGoHome():
      print("No\n")
    else:
      print("Yes\n")

  elif inp == 5:
    if Main.canDoctorGoHome():
      Main.cancelAll()
      print("All appointments cancelled\n")
    else:
      print("\nNo patients in the WRM\n")

  elif inp == 6:
    if Main.canDoctorGoHome():
      Main.ReverseTheLine()
      print("Appointments reversed\n")
    else:
      print("\nNo patients in the WRM\n")

  elif inp == 7:
    print("Thank you for using Waiting Room Management System")
    break
  else:
    print("No such Option")
```
